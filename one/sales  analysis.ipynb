# #i will be using  04 sales data.xlsx  

```python
import pandas as pd
df=pd.read_excel(path+"/04 sales data.xlsx")
df.head()
```

# creating a city column from  Purchase address


```python
df['city']=df['Purchase Address'].astype(str).apply(lambda x:x.split(",")[1:2])
df.to_excel('04 sales data.xlsx',index=False)
df['City']=df['city'].astype(str).apply(lambda x:x.split("[")[1])
df.head()
df.to_excel('04 sales data.xlsx',index=False)
df['CITY']=df['City'].astype(str).apply(lambda x:x.split("'")[1])
df.head()
df.to_excel('04 sales data.xlsx',index=False)
```

# Bellow is the dataframe after adding a new column city

```python
df.head()
```

# Droping unnecesary columns

```python
df=df.drop(['Order Date','Purchase Address','City','city'],axis=1)
df.to_excel('04 sales data.xlsx',index=False)
```

# Adding a Totals column

```python
df['Total']=df['price']*df['quantity_ordered']
df.to_excel('04 sales data.xlsx',index=False)
df.head()
```
