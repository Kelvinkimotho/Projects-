```python


#Started by importing modules

import pandas as pd
```

```python


lines_list_copy=[]

#Opening the csv file using file open function in python

with open("vehicle_info.csv") as target_file:
    #lines list/stores lines after reading them from the csv file
    lines=target_file.readlines()
    #looping through the list of lines getting each line
    for line in lines:
        lines_list_copy.append(line)
        print(line)





```

```python
#Now lets remove the " and replace them with nothing " "
newer_lines_list=[]
for line in lines_list_copy:
    #replacing " with nothing
    newerline=line.replace('"',"")
    newer_lines_list.append(newerline)
    print(newerline)


```

```python


for line in newer_lines_list:
#saving the cleaner lines into another csv file named "cleaner_vehicles_data.csv"
     with open("cleaner_vehicles_data.csv",'a') as cleaner_file:
            cleaner_file.write(f"{line}\n")
            cleaner_file.close()
```

```python



#Using pandas library , i opened the new csv file for the next part of the task  

df=pd.read_csv("cleaner_vehicles_data.csv")
df.head()
```

```python
#Droping the exixting brand name and generating a new one from the vehicle_name column
df=df.drop(['brand_name'],axis=1)
df.to_csv("cleaner_vehicles_data.csv",index=False)
```

```python


#checking whether the column was droped succesfully 
df.head()
```

```python



#Creating a new  brand column  from the  the vehicle name column
df['Brand']=df['vehicle_name'].astype(str).apply(lambda x:x.split(" ")[0])
#Saving the new data in an excel file
df.to_excel("Vehicles_data.xlsx",index=False)
```
