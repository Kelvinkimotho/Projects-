
# Importing pandas module

```python
import pandas as pd
import matplotlib.pyplot as plt
#import os
```

# analysing playstation production data
#Later ploting a  graph of Playstation genres with their number of number of games

```python
df=pd.read_excel('List of best-selling PlayStation 4 video games.xlsx')
genres=df['Genres'].unique()
genre_total_movies_list=[]
for genre in genres:
    genre_total_movies=f"{genre}_total_movies"
    genre_total_movies=0
    
   # print(genre_total_movies)
    for i in range(len(df)):
        if df.loc[i].Genres==genre:
            genre_total_movies+=1
    genre_total_movies_list.append(genre_total_movies) 
    #print(f"{genre},{genre_total_movies}")
   

            
        
```

# A piechart visualizing the relationship between genres and number of playstations in each and every genre

```python
sizes=genre_total_movies_list
labels=genres
plt.pie(sizes,labels=labels,autopct="%1.f%%",counterclock=False)
plt.show()
```
