```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
df=pd.read_csv("Sales.csv")
df.head(4)
```

# task 1: Which country has the most female customers?, plot a bargraph to visualize your answer

```python
#countries lit
countries=df['Country'].unique()
#female customers list per country
female_customers_list=[]
#looping through countries list
for country in countries:
    country_female_customers=f"{country}_female_customers"
    country_female_customers=0
    #looping through the entire dataframe
    for i in range(len(df)):
        if df.loc[i].Customer_Gender=="F" and df.loc[i].Country==country:
            country_female_customers+=1
    female_customers_list.append(country_female_customers)
    print(f"{country} female customers: {country_female_customers}")            
```

```python
#ploting a bargraph
plt.figure(figsize=(8,5))
plt.bar(countries,female_customers_list,color="blue",width=0.5)
plt.xlabel("country")
plt.ylabel("female cuatomers")
plt.title("Female customers per a country")
```

# Task 2: Which country has the most male customers?. 

```python
country_male_customers_list=[]
#looping through the countries list
for country in countries:
    country_male_customers=f"{country}_male_customers"
    country_male_customers=0
    #looping through the entire dataframe
    for j in range(len(df)):
        if df.loc[j].Customer_Gender=="M" and df.loc[j].Country==country:
            country_male_customers+=1
    country_male_customers_list.append(country_male_customers)
    print(f"{country} male customers: {country_male_customers}")
```

```python
#ploting a bargarph 
plt.figure(figsize=(8,5))
plt.bar(countries,country_male_customers_list,color="green",width=0.5)
plt.xlabel("country")
plt.ylabel("Male customers")
plt.title("Male customers per a country")
```

```python
#Modifying Age_Group column
df['Age_Group']=df['Age_Group'].astype(str).apply(lambda x:x.split(" ")[0])
df.to_csv("Sales.csv",index=False)
```

# Task 3:  which Age Group comprises the most Customers?

```python
#Age groups list
age_groups=df['Age_Group'].unique()
#age group customers list
age_group_customers_list=[]
#looping through the age groups
for age_group in age_groups:
    age_group_customers=f"{age_group}_total_customers"
    age_group_customers=0
    for i in range(len(df)):
        if df.loc[i].Age_Group==age_group:
            age_group_customers+=1
    age_group_customers_list.append(age_group_customers)
    print(f"{age_group} customers: {age_group_customers}")
    
    
```

```python
#ploting a bargaph
plt.figure(figsize=(8,5))
plt.barh(age_groups,age_group_customers_list,color="blue")
plt.xlabel("Customers")
plt.ylabel("Age Group")
plt.title("Customers per Age Group")
```
