```python
#importing modules
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
#reading from our dataframe
df=pd.read_csv("Salesstore.csv")
df.head(3)
```

#    Q1.show which shipping mode is most prefered as per total shipments made via  each shipping mode

```python
ship_modes=df['Ship_Mode'].unique()
```

```python
ship_mode_shipments_list=[]
for ship_mode in ship_modes:
    ship_mode_total_shipments=f"{ship_mode}_total_shipments"
    ship_mode_total_shipments=0
    for i in range(len(df)):
        if df.loc[i].Ship_Mode==ship_mode:
            ship_mode_total_shipments+=1
    print(f"{ship_mode} : {ship_mode_total_shipments}")
    ship_mode_shipments_list.append(ship_mode_total_shipments)
```

```python
#visualizing the results using a piechart
sizes=ship_mode_shipments_list
plt.title("shiping modes and  their total shipments")
plt.pie(sizes,labels=ship_modes,autopct="%1.f%%",counterclock=False)
plt.show()
```

# Q2.Which product category had the most sales?

```python
#getting product categories
product_categories=df['Product_Category'].unique()
```

```python
product_category_total_list=[]
#looping through product categorie list
for product_category in product_categories:
    product_category_total_sales=f"{product_category}_total_sales"
    product_category_total_sales=0
    #looping through the dataframe
    for i in range(len(df)):
        if df.loc[i].Product_Category==product_category:
            product_category_total_sales+=df.loc[i].Sales.round()
    product_category_total_list.append(product_category_total_sales)
    print(f"{product_category} : {product_category_total_sales}")
```

```python
#ploting a pierchart 
sizes=product_category_total_list
plt.title("product categories and their total sales percentages ")
plt.pie(sizes,labels=product_categories,autopct="%1.f%%",counterclock=False)
plt.show()
```

# Q3.Which customer segment has the most customers?

```python
#customer segments
customer_segments=df['Customer_Segment'].unique()
#looping through customer segments list
customers_segment_cutomers_list=[]
for customer_segment in customer_segments:
    customer_segment_customers=f"{customer_segment}_customers"
    customer_segment_customers=0
    for i in range(len(df)):
        if  df.loc[i].Customer_Segment==customer_segment:
            customer_segment_customers+=1
    customers_segment_cutomers_list.append(customer_segment_customers)
    print(f"{customer_segment}: {customer_segment_customers}")            
    
```

```python
#ploting a pie chart of customer segments together with their customers
sizes=customers_segment_cutomers_list
plt.title("customer  segments with their customer percentages")
plt.pie(sizes,labels=customer_segments,autopct="%1.f%%",counterclock=False)
plt.show()
```
